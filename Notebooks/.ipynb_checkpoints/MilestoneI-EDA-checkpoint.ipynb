```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# set cut off dates for all of the power related
cutofftop_date = '2013-12-19 00:00:00'
cutoffbot_date = '2021-01-01 00:00:00'
```

```python
def transform_lmp_csv(file_path):
    # Read in csv
    df = pd.read_csv(file_path)
    
    # Drop useless columns
    df.drop(['VALUE', 'TYPE'], axis=1, inplace=True)
    
    # Melt the DF to create hour column then reset index rename columns so everything is in correct format
    df = df.melt(['MARKET_DAY', 'NODE']).set_index(['MARKET_DAY', 'variable']).sort_values(['MARKET_DAY', 'variable']).reset_index().rename({'MARKET_DAY': 'date', 'variable': 'hour'}, axis=1)
    
    # Remove HE from hour and change to datetime
    df['hour'] = df['hour'].str.replace("HE", "").astype('timedelta64[h]')
    
    # Create time column to match weather data and smash the date an hour columns together
    df['time'] = pd.to_datetime(df['date']) + df['hour']
    
    # Set index on time to get it to left side, remove date and hour columns and then reset index
    df = df.set_index('time').drop(['date', 'hour'], axis=1).reset_index()
    
    return df
```

```python
# Various Wind Calculations
#   Change wind speed and direction to U and V components
#   Change U and V components to direction
#   Change U and V components to speed
#   Calculate the angle between two vecors [u1,v1] and [u2,v2]
#  source: https://github.com/blaylockbk/Ute_WRF/blob/master/functions/wind_calcs.py 
def wind_spddir_to_uv(wspd,wdir):
    rad = 4.0*np.arctan(1)/180.
    u = -wspd*np.sin(rad*wdir)
    v = -wspd*np.cos(rad*wdir)

    return u,v

def wind_uv_to_dir(U,V):
    WDIR= (270-np.rad2deg(np.arctan2(V,U)))%360
    return WDIR
    
def wind_uv_to_spd(U,V):
    """
    Calculates the wind speed from the u and v wind components
    Inputs:
      U = west/east direction (wind from the west is positive, from the east is negative)
      V = south/noth direction (wind from the south is positive, from the north is negative)
    """
    WSPD = np.sqrt(np.square(U)+np.square(V))
    return WSPD
    
    
# Below is used for calculing the angle between two wind vectors   
def unit_vector(vector):
    """ Returns the unit vector of the vector.  """
    return vector / np.linalg.norm(vector)

def angle_between(v1, v2):
    """
    Calcualates the angle between two wind vecotrs. Utilizes the cos equation:
                cos(theta) = (u dot v)/(magnitude(u) dot magnitude(v))
                
    Input:
        v1 = vector 1. A numpy array, list, or tuple with 
             u in the first index and v in the second --> vector1 = [u1,v1]
        v2 = vector 2. A numpy array, list, or tuple with 
             u in the first index and v in the second --> vector2 = [u2,v2]
    Output:     
    Returns the angle in radians between vectors 'v1' and 'v2'::
            >>> angle_between((1, 0, 0), (0, 1, 0))
            1.5707963267948966
            >>> angle_between((1, 0, 0), (1, 0, 0))
            0.0
            >>> angle_between((1, 0, 0), (-1, 0, 0))
            3.141592653589793
    """
    v1_u = unit_vector(v1)
    v2_u = unit_vector(v2)
    angle = np.arccos(np.dot(v1_u, v2_u))
    if np.isnan(angle):
        if (v1_u == v2_u).all():
            return np.rad2deg(0.0)
        else:
            return np.rad2deg(np.pi)
    return np.rad2deg(angle)
```

# Are those high prices during low precipitation in the summer when it's also hot?

```python
import datetime as dt

def get_season(df_date):
#     print(df_date)
    date = df_date # .now()
#     print(date)
    month_day = date.month * 100 + date.day
#     print(month_day)

    if ((month_day > 320) and (month_day < 621)):
        season = 'spring'
    elif ((month_day > 620) and (month_day < 923)):
        season = 'summer'
    elif ((month_day > 922) and (month_day < 1223)):
        season = 'fall'
    else:
        season = 'winter' 
        
    return season
```

```python
mi_data_for_plot_wtime = mi_finaldf.copy()

seasons = []
for index, row in mi_data_for_plot_wtime.iterrows():
    
    date_time_str = str(mi_data_for_plot_wtime['time'][index].to_pydatetime())
    date_time_obj = dt.datetime.strptime(date_time_str, '%Y-%m-%d %H:%M:%S')
    
    season = get_season(date_time_obj)
    seasons.append(season) 

mi_data_for_plot_wtime['season'] = seasons

mi_data_for_plot_wtime = mi_data_for_plot_wtime.drop(columns=['time']) 
mi_data_for_plot_wtime.sample(n=10, random_state=1)
```

```python
#plot important boxes
sns.pairplot(
    mi_data_for_plot_wtime, 
    hue="season", 
    x_vars=['det_surf_pres','det_wind_mph','det_temp_f','det_precip_in'],
    y_vars=['mi_ra_price', 'mi_da_price'],
);
```
