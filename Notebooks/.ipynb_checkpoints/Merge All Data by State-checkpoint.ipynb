```python
import pandas as pd
import os
```

# Michigan

```python
mi_weather = pd.read_csv(r'../data/weather/mi_weather_full.csv')
```

```python
mi_lmp = pd.read_csv(r'../data/lmp/MI/mi_lmp.csv').drop('NODE', axis=1)
```

```python
mi_load = pd.read_csv(r'../data/load/load.csv', usecols=['time', 'MI_load_forecast', 'MI_load_actual'])
```

```python
mi = mi_weather.merge(mi_lmp, on='time').merge(mi_load, on='time')
mi.to_csv(r'../data/mi_full.csv', index=False)
mi
```

# Louisiana

```python
la_weather = pd.read_csv(r'../data/weather/la_weather_full.csv')
```

```python
la_lmp = pd.read_csv(r'../data/lmp/LA/la_lmp.csv').drop('NODE', axis=1)
```

```python
la_load = pd.read_csv(r'../data/load/load.csv', usecols=['time', 'LA_load_forecast', 'LA_load_actual'])
```

```python
la = la_weather.merge(la_lmp, on='time').merge(la_load, on='time')
la.to_csv(r'../data/la_full.csv', index=False)
la
```

# Full

```python
full = mi.merge(la, on='time')
full.to_csv(r'../data/energy_data.csv', index=False)
full
```
